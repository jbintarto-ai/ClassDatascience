## Simulasi Sederhana NLP: Part-of-Speech Tagging (POS Tagging)

**Part-of-Speech Tagging (POS Tagging)** adalah proses memberi label pada setiap kata dalam teks dengan kategori gramatikalnya (misalnya, kata benda, kata kerja, kata sifat, adverbia, dll.). Ini adalah langkah penting dalam banyak aplikasi NLP karena membantu memahami struktur dan makna kalimat.

### Mengapa POS Tagging Penting?

*   **Word Sense Disambiguation:** Membedakan makna kata yang sama dengan kategori gramatikal yang berbeda (misal: "bank" sebagai institusi finansial vs. "bank" sebagai tepi sungai).
*   **Information Extraction:** Membantu mengidentifikasi jenis entitas dalam teks (misal: kata benda seringkali merupakan nama entitas).
*   **Machine Translation:** Membantu dalam penerjemahan dengan memahami peran gramatikal kata dalam kalimat.
*   **Text Analysis:** Memberikan wawasan tentang struktur bahasa yang digunakan dalam sebuah dokumen.

### Contoh Tag POS Sederhana (dalam Bahasa Inggris):

*   **NN:** Noun, singular (kata benda tunggal)
*   **NNS:** Noun, plural (kata benda jamak)
*   **VB:** Verb, base form (kata kerja bentuk dasar)
*   **VBD:** Verb, past tense (kata kerja bentuk lampau)
*   **JJ:** Adjective (kata sifat)
*   **RB:** Adverb (kata keterangan)
*   **DT:** Determiner (penentu, misal: 'the', 'a')

Untuk bahasa Indonesia, ada set tag yang berbeda, dan seringkali membutuhkan model yang dilatih khusus.

```python
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import nltk

# Ensure the necessary NLTK resources are downloaded for tokenization and POS tagging
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True) # Added to fix the LookupError

# Teks contoh (dalam Bahasa Inggris karena tagger default NLTK berbahasa Inggris)
english_text = "The quick brown fox jumps over the lazy dog."

print(f"Teks Asli: {english_text}\n")

# 1. Tokenisasi
tokens_en = word_tokenize(english_text)
print(f"Token Bahasa Inggris: {tokens_en}\n")

# 2. POS Tagging
pos_tags_en = pos_tag(tokens_en)
print(f"POS Tags Bahasa Inggris: {pos_tags_en}")
```
