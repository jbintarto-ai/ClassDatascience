# Apa itu Plotly
> Plotly adalah pustaka (library) visualisasi data sumber terbuka (open-source) yang sangat populer digunakan untuk membuat grafik yang interaktif, dekoratif, dan siap pakai untuk web.
Tidak seperti pustaka visualisasi statis seperti Matplotlib atau Seaborn di Python, Plotly memungkinkan pengguna untuk memperbesar (zoom), menggeser (pan), melihat detail data saat kursor diarahkan ke grafik (hover), hingga menyembunyikan atau menampilkan data tertentu dengan sekali klik.
1. Keunggulan Utama Plotly
Sangat Interaktif: Grafik yang dihasilkan bukan sekadar gambar mati. Pengguna bisa berinteraksi langsung dengan data di dalam grafik.
Multi-Bahasa Pemrograman: Meskipun sangat terkenal di ekosistem Python, Plotly juga tersedia secara native untuk R, JavaScript (Plotly.js), MATLAB, dan Julia.
Mendukung Berbagai Jenis Grafik: Mulai dari grafik standar (batang, garis, pai), grafik statistik (box plot, histogram), grafik finansial (candlestick), hingga visualisasi 3D dan peta (maps/geospatial).
Plotly Express: Ini adalah interface tingkat tinggi (high-level) dari Plotly yang memungkinkan kamu membuat grafik kompleks hanya dengan beberapa baris kode saja.
2. Hubungan Plotly dengan Dash
Plotly sering kali disandingkan dengan Dash. Dash adalah kerangka kerja (framework) buatan Plotly yang digunakan untuk membangun aplikasi web analitik dan dashboard interaktif secara penuh hanya dengan menggunakan kode Python (tanpa perlu mahir HTML, CSS, atau JavaScript).

# Jenis plotly
1. Plotly Express (plotly.express)

  * Plotly Express (sering disingkat px) adalah wrapper tingkat tinggi (high-level interface) yang dirancang agar pengguna bisa membuat grafik interaktif dengan sangat cepat.

  * Karakteristik: Hanya butuh satu atau beberapa baris kode saja. Sangat cocok digunakan bersama dengan Pandas DataFrame.

  * Kelebihan: Otomatis mengatur warna, legenda, hover label, dan konfigurasi dasar grafik secara cerdas.

  * Kapan digunakan? Untuk eksplorasi data sehari-hari (Exploratory Data Analysis) atau saat Anda ingin membuat grafik standar (seperti scatter, bar, line, box plot) tanpa mau pusing dengan kustomisasi yang rumit.

2. Plotly Graph Objects (plotly.graph_objects)

* Plotly Graph Objects (sering disingkat go) adalah modul tingkat rendah (low-level interface) yang menjadi fondasi dari seluruh grafik di Plotly.

* Karakteristik: Struktur kodenya lebih panjang dan eksplisit karena setiap elemen grafik (sumbu, warna, titik data, layout) didefinisikan secara manual dalam bentuk dictionary atau objek Python.

* Kelebihan: Memberikan kontrol 100% terhadap setiap detail visualisasi Anda. Anda bisa menggabungkan berbagai jenis grafik yang tidak biasa dalam satu kanvas.

* Kapan digunakan? Ketika Plotly Express tidak lagi mampu memenuhi kebutuhan desain Anda yang sangat spesifik atau saat Anda membuat grafik kustom yang rumit.

3. Plotly Subplots (plotly.subplots)
* Ini adalah modul khusus yang digunakan untuk menggabungkan beberapa grafik (plots) ke dalam satu tampilan grid atau panel (seperti susunan 2x2 atau 1x3).

* Karakteristik: Menggunakan fungsi make_subplots untuk membagi area kanvas menjadi beberapa baris dan kolom.

* Kapan digunakan? Saat Anda ingin membandingkan beberapa tren data atau jenis grafik yang berbeda secara berdampingan dalam satu waktu.

4. Dash oleh Plotly
* Meskipun Dash adalah framework terpisah, ia merupakan bagian tak terpisahkan dari ekosistem Plotly untuk Python.

* Karakteristik: Dash digunakan untuk mengubah grafik-grafik Plotly yang sudah Anda buat menjadi aplikasi web atau dashboard analitik yang utuh dan interaktif tanpa perlu menguasai HTML, CSS, atau JavaScript.

* Kapan digunakan? Saat Anda ingin membagikan hasil analisis Anda kepada tim atau klien dalam bentuk aplikasi web interaktif yang memiliki tombol filter, dropdown, dan input dinamis lainnya.

## Analisis Diagnosa Penyakit dengan Plotly

Untuk menganalisis diagnosa penyakit, kita akan mengikuti langkah-langkah berikut:

1.  **Membuat Data Sintetis**: Karena tidak ada data yang disediakan, kita akan membuat dataset tiruan yang berisi informasi pasien, gejala, dan diagnosis.
2.  **Eksplorasi Data Awal**: Melihat sekilas struktur data dan distribusi diagnosis.
3.  **Visualisasi Interaktif dengan Plotly**: Membuat grafik interaktif untuk memahami hubungan antara gejala dan diagnosis.

### Pembaruan Interaktivitas:

Saya telah memodifikasi grafik-grafik di bawah ini untuk menjadi lebih interaktif dengan menambahkan faset (`facet_col='gender'`). Ini berarti Anda dapat melihat setiap grafik dipecah menjadi dua kolom terpisah: satu untuk pria dan satu untuk wanita, memungkinkan perbandingan yang lebih mudah dan wawasan yang lebih mendalam mengenai bagaimana diagnosis, gejala, dan usia bervariasi berdasarkan jenis kelamin.

```python
# Import library yang diperlukan
import pandas as pd
import numpy as np
import plotly.express as px
import warnings

# Menyembunyikan FutureWarning dari Plotly Express
warnings.filterwarnings('ignore', category=FutureWarning)

# 1. Membuat Data Sintetis untuk Diagnosa Penyakit
np.random.seed(42)
num_patients = 500

data = {
    'patient_id': range(1, num_patients + 1),
    'age': np.random.randint(18, 70, num_patients),
    'gender': np.random.choice(['Male', 'Female'], num_patients, p=[0.5, 0.5]),
    'symptom_fever': np.random.choice([0, 1], num_patients, p=[0.6, 0.4]),
    'symptom_cough': np.random.choice([0, 1], num_patients, p=[0.55, 0.45]),
    'symptom_sore_throat': np.random.choice([0, 1], num_patients, p=[0.7, 0.3]),
    'symptom_fatigue': np.random.choice([0, 1], num_patients, p=[0.5, 0.5]),
    'symptom_headache': np.random.choice([0, 1], num_patients, p=[0.65, 0.35]),
    'diagnosis': np.random.choice(['Flu', 'Common Cold', 'Bronchitis', 'Allergy'], num_patients, p=[0.3, 0.4, 0.2, 0.1])
}

df = pd.DataFrame(data)

# Menyesuaikan diagnosis berdasarkan gejala
def assign_diagnosis(row):
    if row['symptom_fever'] == 1 and row['symptom_cough'] == 1:
        if row['symptom_sore_throat'] == 1:
            return 'Flu'
        else:
            return 'Bronchitis'
    elif row['symptom_cough'] == 1 and row['symptom_sore_throat'] == 1:
        return 'Common Cold'
    elif row['symptom_fatigue'] == 1 and row['symptom_headache'] == 1:
        return 'Flu'
    else:
        return row['diagnosis'] # Keep original if no specific pattern

df['diagnosis'] = df.apply(assign_diagnosis, axis=1)

# Mengganti nama kolom ke Bahasa Indonesia
df = df.rename(columns={
    'patient_id': 'id_pasien',
    'age': 'usia',
    'gender': 'jenis_kelamin',
    'symptom_fever': 'gejala_demam',
    'symptom_cough': 'gejala_batuk',
    'symptom_sore_throat': 'gejala_sakit_tenggorokan',
    'symptom_fatigue': 'gejala_kelelahan',
    'symptom_headache': 'gejala_sakit_kepala'
})

# Tampilkan 5 baris pertama data dengan nama kolom yang diperbarui
print("Data Diagnosa Penyakit Sintetis (dengan nama kolom Bahasa Indonesia):")
display(df.head())
```

### Eksplorasi Data Awal: Distribusi Diagnosis

```python
# Menghitung frekuensi diagnosis berdasarkan jenis kelamin
diagnosis_gender_freq = df.groupby(['jenis_kelamin', 'diagnosis']).size().unstack(fill_value=0)

print("Tabel Frekuensi Diagnosis berdasarkan Jenis Kelamin:")
display(diagnosis_gender_freq)
```

```python
diagnosis_gender_freq_melted = diagnosis_gender_freq.reset_index().melt(id_vars='jenis_kelamin', var_name='Diagnosis', value_name='Jumlah Pasien')

fig = px.bar(
    diagnosis_gender_freq_melted,
    x='jenis_kelamin',
    y='Jumlah Pasien',
    color='Diagnosis',
    barmode='group',
    title='Distribusi Diagnosis berdasarkan Jenis Kelamin',
    labels={
        'jenis_kelamin': 'Jenis Kelamin',
        'Jumlah Pasien': 'Jumlah Pasien',
        'Diagnosis': 'Diagnosis'
    },
    template='plotly_white'
)
fig.update_layout(
    xaxis_title_text='Jenis Kelamin',
    yaxis_title_text='Jumlah Pasien',
    legend_title_text='Diagnosis'
)
fig.show()
```

Visualisasi Hubungan Gejala dan Diagnosis

```python
# Agregasi data untuk melihat prevalensi gejala per diagnosis
symptom_diagnosis_melted = df.melt(id_vars=['id_pasien', 'diagnosis', 'jenis_kelamin'],
                                 value_vars=['gejala_demam', 'gejala_batuk', 'gejala_sakit_tenggorokan', 'gejala_kelelahan', 'gejala_sakit_kepala'],
                                 var_name='Gejala', value_name='Ada')

# Filter hanya gejala yang 'Ada' (nilai 1)
symptom_diagnosis_present = symptom_diagnosis_melted[symptom_diagnosis_melted['Ada'] == 1]

symptom_counts_per_diagnosis = symptom_diagnosis_present.groupby(['diagnosis', 'Gejala', 'jenis_kelamin']).size().reset_index(name='Jumlah')
display(symptom_counts_per_diagnosis)
# Buat grouped bar chart interaktif
# Menambahkan facet_col='gender' untuk memecah grafik berdasarkan jenis kelamin
fig = px.bar(
    symptom_counts_per_diagnosis,
    x='Gejala',
    y='Jumlah',
    color='diagnosis',
    barmode='group',
    title='Prevalensi Gejala per Diagnosa berdasarkan Jenis Kelamin',
    labels={'Gejala': 'Gejala', 'Jumlah': 'Jumlah Pasien dengan Gejala', 'jenis_kelamin': 'Jenis Kelamin'},
    template='plotly_white',
    facet_col='jenis_kelamin' # Tambahan untuk interaktivitas
)
fig.update_layout(
    xaxis_title_text='Gejala',
    yaxis_title_text='Jumlah Pasien',
    legend_title_text='Diagnosis'
)
fig.show()
```

### Distribusi Usia per Diagnosa

```python
import pandas as pd

# Hitung statistik deskriptif untuk 'usia' berdasarkan 'diagnosis' dan 'jenis_kelamin'
usia_diagnosis_gender_summary = df.groupby(['diagnosis', 'jenis_kelamin'])['usia'].describe()

print("Tabel Ringkasan Distribusi Usia per Diagnosa berdasarkan Jenis Kelamin:")
display(usia_diagnosis_gender_summary)
```

```python
# Buat box plot interaktif untuk distribusi usia per diagnosa
# Menambahkan facet_col='jenis_kelamin' untuk memecah grafik berdasarkan jenis kelamin
fig = px.box(
    df,
    x='diagnosis',
    y='usia',
    color='diagnosis',
    title='Distribusi Usia per Diagnosa berdasarkan Jenis Kelamin',
    labels={'diagnosis': 'Diagnosis Penyakit', 'usia': 'Usia', 'jenis_kelamin': 'Jenis Kelamin'},
    template='plotly_white',
    facet_col='jenis_kelamin' # Tambahan untuk interaktivitas
)
fig.update_layout(
    xaxis_title_text='Diagnosis Penyakit',
    yaxis_title_text='Usia',
    showlegend=False
)
fig.show()
```

### Fitur Pilihan Interaktif Baru: Grafik Sunburst

Grafik sunburst di bawah ini menyediakan cara interaktif untuk menjelajahi distribusi pasien secara hierarkis. Anda dapat mengklik pada segmen (misalnya, diagnosis tertentu) untuk "memperbesar" dan melihat distribusinya berdasarkan jenis kelamin, lalu mengklik lagi untuk melihat distribusi berdasarkan kelompok usia dalam kategori tersebut.

```python
# Membuat kelompok usia untuk analisis lebih lanjut
bins = [18, 30, 45, 60, 70]
labels = ['18-29', '30-44', '45-59', '60-70']
df['kelompok_usia'] = pd.cut(df['usia'], bins=bins, labels=labels, right=False)

# Buat sunburst chart interaktif
fig = px.sunburst(
    df,
    path=['diagnosis', 'jenis_kelamin', 'kelompok_usia'], # Hierarki untuk sunburst
    values='id_pasien', # Nilai yang dihitung
    title='Distribusi Pasien Berdasarkan Diagnosis, Jenis Kelamin, dan Kelompok Usia',
    template='plotly_white',
    color='diagnosis' # Warna berdasarkan diagnosis utama
)
fig.update_traces(hovertemplate='<b>%{label}</b> <br>Jumlah Pasien: %{value}<br>Persentase: %{percentParent:.2%}')
fig.show()
```

# Peran plotly dalam membuat model ML
kasus : prediksi diagnosa penyakit jantung

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
import plotly.express as px
import plotly.graph_objects as go

# 1. Membuat dataset sintetis untuk prediksi penyakit jantung
np.random.seed(42)
num_samples = 1000

data_ml = {
    'age': np.random.randint(30, 75, num_samples),
    'sex': np.random.choice([0, 1], num_samples, p=[0.4, 0.6]), # 0=Female, 1=Male
    'cholesterol': np.random.randint(150, 300, num_samples),
    'blood_pressure': np.random.randint(100, 180, num_samples),
    'max_hr': np.random.randint(120, 200, num_samples),
    'exercise_angina': np.random.choice([0, 1], num_samples, p=[0.7, 0.3]), # 0=No, 1=Yes
}

df_ml = pd.DataFrame(data_ml)

# Membuat variabel target 'heart_disease' berdasarkan aturan yang disimulasikan
# Misal: usia lebih tinggi, kolesterol, tekanan darah, dan angina saat berolahraga meningkatkan risiko
heart_disease_risk = (
    0.3 * (df_ml['age'] / df_ml['age'].max()) +
    0.2 * df_ml['sex'] +
    0.25 * (df_ml['cholesterol'] / df_ml['cholesterol'].max()) +
    0.2 * (df_ml['blood_pressure'] / df_ml['blood_pressure'].max()) +
    0.1 * df_ml['exercise_angina'] -
    0.15 * (df_ml['max_hr'] / df_ml['max_hr'].max()) # Denyut jantung maksimal yang lebih tinggi mungkin menurunkan risiko
)
df_ml['heart_disease'] = (heart_disease_risk > np.percentile(heart_disease_risk, 60)).astype(int) # 40% teratas memiliki penyakit jantung

# Mengganti nama kolom ke Bahasa Indonesia
df_ml = df_ml.rename(columns={
    'age': 'usia',
    'sex': 'jenis_kelamin',
    'cholesterol': 'kolesterol',
    'blood_pressure': 'tekanan_darah',
    'max_hr': 'detak_jantung_maks',
    'exercise_angina': 'angina_saat_olahraga',
    'heart_disease': 'penyakit_jantung'
})

print("Dataset Sintetis Penyakit Jantung :")
display(df_ml.head())

# 2. Membagi data menjadi training dan testing sets
X = df_ml.drop('penyakit_jantung', axis=1)
y = df_ml['penyakit_jantung']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 3. Melatih model Machine Learning sederhana (Regresi Logistik)
model = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)
model.fit(X_train, y_train)

# 4. Melakukan prediksi pada test set
y_pred = model.predict(X_test)

# 5. Mengevaluasi model - Confusion Matrix dan Akurasi
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)

print(f"\nAkurasi Model: {accuracy:.2f}")
print("Confusion Matrix:")
print(cm)

# 6. Memvisualisasikan Confusion Matrix menggunakan Plotly
cm_labels = ['Tidak Sakit Jantung', 'Sakit Jantung']

fig_ml_cm = px.imshow(
    cm,
    x=cm_labels,
    y=cm_labels,
    color_continuous_scale='Blues',
    text_auto=True,
    labels=dict(x="Prediksi Model", y="Kondisi Aktual Pasien"),
    title="<b>Confusion Matrix Prediksi Penyakit Jantung</b>"
)

fig_ml_cm.update_layout(
    xaxis_title="Hasil Prediksi Model",
    yaxis_title="Kondisi Nyata Pasien"
)

fig_ml_cm.show()

# Memvisualisasikan Koefisien Fitur (Feature Importance) untuk Regresi Logistik
feature_importance = pd.DataFrame({
    'Fitur': X.columns,
    'Koefisien': model.coef_[0]
}).sort_values(by='Koefisien', ascending=False)

fig_feature_imp = px.bar(
    feature_importance,
    x='Fitur',
    y='Koefisien',
    color='Koefisien',
    color_continuous_scale=px.colors.sequential.RdBu,
    title='<b>Koefisien Fitur Model Regresi Logistik (Pengaruh pada Prediksi Sakit Jantung)</b>',
    labels={'Koefisien': 'Pengaruh pada Prediksi Sakit Jantung'}
)
fig_feature_imp.update_layout(showlegend=False)
fig_feature_imp.show()
```

### Prediksi Diagnosis Penyakit Jantung untuk Pasien Baru

Sekarang, setelah model dilatih, kita bisa menggunakan model tersebut untuk memprediksi diagnosis penyakit jantung untuk data pasien baru. Anda bisa mengubah nilai-nilai input di bawah untuk melihat bagaimana model membuat prediksi.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

def predict_new_patient_heart_disease(model, patient_data: dict, feature_columns: list) -> str:
    # Pastikan data pasien baru memiliki kolom yang sama dengan data pelatihan
    new_patient_df = pd.DataFrame([patient_data], columns=feature_columns)

    # Lakukan prediksi
    prediction = model.predict(new_patient_df)
    prediction_proba = model.predict_proba(new_patient_df)

    result = ""
    if prediction[0] == 1:
        result = f"Pasien diprediksi memiliki **Penyakit Jantung** (Probabilitas: {prediction_proba[0][1]*100:.2f}%)"
    else:
        result = f"Pasien diprediksi **Tidak Memiliki Penyakit Jantung** (Probabilitas: {prediction_proba[0][0]*100:.2f}%)"

    return result

# --- Re-defining dependencies from previous cell for independent execution ---
# 1. Membuat dataset sintetis untuk prediksi penyakit jantung
np.random.seed(42)
num_samples = 1000

data_ml = {
    'age': np.random.randint(30, 75, num_samples),
    'sex': np.random.choice([0, 1], num_samples, p=[0.4, 0.6]), # 0=Female, 1=Male
    'cholesterol': np.random.randint(150, 300, num_samples),
    'blood_pressure': np.random.randint(100, 180, num_samples),
    'max_hr': np.random.randint(120, 200, num_samples),
    'exercise_angina': np.random.choice([0, 1], num_samples, p=[0.7, 0.3]), # 0=No, 1=Yes
}

df_ml = pd.DataFrame(data_ml)

# Membuat variabel target 'heart_disease' berdasarkan aturan yang disimulasikan
heart_disease_risk = (
    0.3 * (df_ml['age'] / df_ml['age'].max()) +
    0.2 * df_ml['sex'] +
    0.25 * (df_ml['cholesterol'] / df_ml['cholesterol'].max()) +
    0.2 * (df_ml['blood_pressure'] / df_ml['blood_pressure'].max()) +
    0.1 * df_ml['exercise_angina'] -
    0.15 * (df_ml['max_hr'] / df_ml['max_hr'].max())
)
df_ml['heart_disease'] = (heart_disease_risk > np.percentile(heart_disease_risk, 60)).astype(int)

# Mengganti nama kolom ke Bahasa Indonesia
df_ml = df_ml.rename(columns={
    'age': 'usia',
    'sex': 'jenis_kelamin',
    'cholesterol': 'kolesterol',
    'blood_pressure': 'tekanan_darah',
    'max_hr': 'detak_jantung_maks',
    'exercise_angina': 'angina_saat_olahraga',
    'heart_disease': 'penyakit_jantung'
})

# 2. Membagi data menjadi training dan testing sets
X = df_ml.drop('penyakit_jantung', axis=1)
y = df_ml['penyakit_jantung']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# 3. Melatih model Machine Learning sederhana (Regresi Logistik)
model = LogisticRegression(random_state=42, solver='liblinear', max_iter=1000)
model.fit(X_train, y_train)
# --- End of dependencies re-definition ---

# Contoh data pasien baru
# Mengganti nama fitur ke Bahasa Indonesia agar sesuai dengan X.columns
new_patient = {
    'usia': 65,
    'jenis_kelamin': 1, # 1=Male
    'kolesterol': 280,
    'tekanan_darah': 160,
    'detak_jantung_maks': 140,
    'angina_saat_olahraga': 1 # 1=Yes
}

# Dapatkan nama kolom fitur dari data pelatihan (X)
feature_columns = X.columns.tolist()

print(f"Data Pasien Baru: {new_patient}")
prediction_result = predict_new_patient_heart_disease(model, new_patient, feature_columns)
print(f"\nHasil Prediksi: {prediction_result}")

# Contoh pasien lain
# Mengganti nama fitur ke Bahasa Indonesia agar sesuai dengan X.columns
new_patient_2 = {
    'usia': 40,
    'jenis_kelamin': 0, # 0=Female
    'kolesterol': 180,
    'tekanan_darah': 120,
    'detak_jantung_maks': 180,
    'angina_saat_olahraga': 0 # 0=No
}

print(f"\nData Pasien Baru 2: {new_patient_2}")
prediction_result_2 = predict_new_patient_heart_disease(model, new_patient_2, feature_columns)
print(f"Hasil Prediksi: {prediction_result_2}")
```
